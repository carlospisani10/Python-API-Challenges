# Pyweather: Weather information for 500 random cities across the world. Citipy was used to find nearest cities from randomly generated Latitudes and Longitudes. Openweathermapy was used to then retrieve weather information.

# Three Takeaways
1. Temperature in the negative latitudes are higher than that of positive latitudes. This is due to the tilt of the earth currenlty facing the sun in the negative latitudes. Hence why positive latitude countries (ex: USA) are approaching winter and negative latitude countries (ex: Argentina) are approaching summer. 
2. Our cities list tend to have minimal windspeed. Note that this is not a definite pattern like takeaway 1 as other factors can affect windspeed.
3. Moisture is on the rise! For our set of cities, the majority are currenly experiencing over 50% humidity levels. This is likely due to rising CO2 levels causing higher temp and humidity levels.

```python
#Imoprt Dependencies
import requests
import matplotlib.pyplot as plt
import requests as req
import pandas as pd
import openweathermapy as ow
import json
from citipy import citipy
import numpy as np
import random
from weather_key2 import api_key
```

```python
#Create a empty list and use loop and citipy to find random cities
lat = []
lng = []
cities = set([])
           
while len(cities) < 1500:
    latitude = (random.randrange(-90,90,1))
    longitude = (random.randrange(-180, 180,1))
    city = citipy.nearest_city(latitude,longitude)
    city = city.city_name
    lat.append(latitude)
    lng.append(longitude)
    cities.add(city)
len(cities)
```

```python
#run a loop in the ow module with our 1500 random cities, create an if parameter to cut off at 500
settings = {"units": "imperial", "appid": api_key}
weather_data = []
for city in cities:
    if len(weather_data) < 500:
        try:
            data = ow.get_current(city, **settings)
            weather_data.append(data)
           
        
        except Exception as e:
            print(e)
            continue

summary = ["name", "coord.lat", "main.temp", "main.humidity", "clouds.all", "wind.speed"]

# Create a Pandas DataFrame with the results
data = [response(*summary) for response in weather_data]

column_names = ["City", "Latitude","Temperature", "Humidity", "Cloud Coverage", "Wind Speed"]
weather_data = pd.DataFrame(data, columns=column_names)
weather_data
```

```python
#export data_frame solution to CSV
weather_data.to_csv("CSV/Solution.csv")
```

```python
#Temperature (F) vs. Latitude
x_axis = weather_data["Latitude"]
data = weather_data["Temperature"]

plt.style.use('seaborn')
plt.scatter(x_axis, data, marker="o", facecolors="red", edgecolors="black", alpha=0.75)
plt.title("Temperature vs Latitude for Random Cities")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.savefig("Pyweather_Graphs/Temp vs Latitude")
plt.show()
```

```python
#Humidity (%) vs. Latitude
x_axis = weather_data["Humidity"]
data = weather_data["Temperature"]

plt.style.use('seaborn')
plt.scatter(x_axis, data, marker="o", facecolors="green", edgecolors="black", alpha=0.75)
plt.title("Humidity vs Latitude for Random Cities")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.savefig("Pyweather_Graphs/Humidity vs Latitude")
plt.show()
```

```python
#Cloudiness (%) vs. Latitude
x_axis = weather_data["Latitude"]
data = weather_data["Cloud Coverage"]

plt.style.use('seaborn')
plt.scatter(x_axis, data, marker="o", facecolors="blue", edgecolors="black", alpha=0.75)
plt.title("Cloudiness vs Latitude for Random Cities")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness (%)")
plt.savefig("Pyweather_Graphs/Cloudiness vs Latitude")
plt.show()
```

```python
#Windspeed (%) vs. Latitude
x_axis = weather_data["Latitude"]
data = weather_data["Wind Speed"]

plt.style.use('seaborn')
plt.scatter(x_axis, data, marker="o", facecolors="magenta", edgecolors="black", alpha=0.75)
plt.title("Windspeed vs Latitude for Random Cities")
plt.xlabel("Latitude")
plt.ylabel("Windspeed (mph)")
plt.savefig("Pyweather_Graphs/Wind Speed vs Latitude")
plt.show()
```
